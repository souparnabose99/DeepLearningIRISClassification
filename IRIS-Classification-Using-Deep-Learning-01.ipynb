# Creating Deep Neural Network for classifying famous IRIS dataset

<b>Importing Dataset<b/>

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
type(iris)
```

<b>IRIS dataset description<b/>

```python
print(iris.DESCR)
```

```python
X = iris.data
```

```python
X
```

```python
y = iris.target
```

```python
y
```

<b>One Hot Encoding<b/>

```python
#class1 = [1,0,0], class2 = [0,1,0], class3 = [0,0,1]
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.shape
```

<b>Splitting data into training & test set<b/>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_test
```

<b>Scaling Data<b/>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_train
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
scaled_X_test
```

<b>Creating Neural Network for model training<b/>

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(8, input_dim = 4, activation = 'relu'))
model.add(Dense(8, input_dim = 4, activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(scaled_X_train, y_train, epochs=150, verbose=2)
```

```python
model.predict(scaled_X_test)
```

<b>Prediction & Metrics reporting<b/>

```python
model.predict_classes(scaled_X_test)
```

```python
predictions = model.predict_classes(scaled_X_test)
```

```python
y_test.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test.argmax(axis=1), predictions))
print(classification_report(y_test.argmax(axis=1), predictions))
```

<b>Saving Model<b/>

```python
model.save('IrisNlpDeep.h5')
```
